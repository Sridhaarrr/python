```python
import numpy as np
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
#data = np.array([10,5,8,9,15,22,26,11,15,16,18,7])
```

```python
for i in range(len(data.iloc[:,2])):    
    historicalData = get_history(symbol = data.iloc[i,2], start=start, end=end)
    historicalData["Company"] = data.iloc[i,0]
    ma = moving_average(historicalData['Close'], 100)
    historicalData["MAV"] = (len(historicalData['Close'])-len(ma))*[0]+list(ma)
    for i, row in historicalData.iterrows():
        if i> datetime.date.today() - timedelta(days = 2) and row['MAV']<row['Close']:
            print(i,row['Company'],row['MAV'],row['Close'])
```
